```python
JAILBREAK_DATAFILE="text_jailbreaks.csv"


def read_jailbreak_data(file_path, sampling_ratio):
    with open(file_path, mode='r', newline='', encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = next(csv_reader)  # Skip the header row
        print("Header:", header)   # Optional: print header row
        i = 0
        for row in csv_reader:
            i = i + 1
            if (random.random() <= sampling_ratio):
                print(row[6])
        print("Number of lines = ", i)
```

```python
read_jailbreak_data(JAILBREAK_DATAFILE, 0.05)
```

```python
from openai import OpenAI

client=OpenAI()

def oai_embedding(input):
    x = client.embeddings.create(
      model="text-embedding-ada-002",
      input=input,
      encoding_format="float")
    return x.data[0].embedding
```

```python
# Compute embeddings for all of the jail break data
embedding_dict = {}

def compute_embeddings_jailbreak_data(file_path, sampling_ratio):
    embedding_dict = {}
    with open(file_path, mode='r', newline='', encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = next(csv_reader)  # Skip the header row
        print("Header:", header)   # Optional: print header row
        i = 0
        for row in csv_reader:
            i = i + 1
            if (random.random() <= sampling_ratio):
                embedding_dict[row[6]] = oai_embedding(row[6])
        print("Number of lines = ", i)
    return embedding_dict

embedding_dict = compute_embeddings_jailbreak_data(JAILBREAK_DATAFILE, 1.1)

```

```python
import pickle 


JB_EMBED_OUTFILE = "jb_embed.pickle"
with open(JB_EMBED_OUTFILE, "wb") as outfile:
    pickle.dump(embedding_dict, outfile)
```

```python
# Load AutoSafe filter
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

NV_EMBED_OUTFILE = "nvidia_embed.pickle"

with open(NV_EMBED_OUTFILE, "rb") as infile: 
    nv_embed = pickle.load(infile)

# print(data_pairs)

def compute_cos(X, Y): 
    # Convert embeddings to a numpy array
    np_X = np.array(X)
    np_Y = np.array(Y)

    # Compute cosine similarity between the embeddings
    similarity_matrix = cosine_similarity(np_X, np_Y)
    return similarity_matrix
    
def autosafe_filter(msg, nv_embed):
    msg_embed = oai_embedding(msg)
    msg_sim = compute_cos(nv_embed, [msg_embed])
    # print(msg_sim)
    return (np.mean(msg_sim[0]) > 0.72)

def autosafe_filter_given_embed(msg_embed, nv_embed):
    msg_sim = compute_cos(nv_embed, [msg_embed])
    # print(msg_sim)
    return (np.mean(msg_sim[0]) > 0.72)
```

```python
true_positive = 0
false_negative = 0
for k, v in embedding_dict.items():
    if autosafe_filter_given_embed(v, nv_embed):
        false_negative = false_negative + 1
    else:
        true_positive = true_positive + 1
print("total = ", len(embedding_dict.items()), " false_negative ", false_negative, " true positive = ", true_positive)
```
